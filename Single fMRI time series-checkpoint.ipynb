```python
# 1. SETUP
import torch
import numpy as np
import matplotlib.pyplot as plt
import vbi
from sbi.inference import SNPE
from sbi.utils import BoxUniform

# ---

# 2. DATA LOADING
sample_data = vbi.LoadSample(nn=88)
all_regions_ts = sample_data.get_bold()
single_region_ts = all_regions_ts[0, :]
observations = torch.from_numpy(single_region_ts).float()

# ---

# 3. DEFINE THE SIMULATOR
def ar1_simulator(theta):
    phi, sigma = theta
    n = len(observations)
    y = torch.zeros(n)
    for t in range(1, n):
        y[t] = phi * y[t - 1] + torch.normal(0.0, sigma, size=(1,))
    return y

# ---

# 4. SETUP AND RUN SBI
# Define a STABLE prior distribution for the two parameters (phi and sigma)
# By keeping phi between -0.99 and 0.99, we prevent unstable simulations.
low_bounds = torch.tensor([-0.99, 0.01])   
high_bounds = torch.tensor([0.99, 2.0])    
prior = BoxUniform(low=low_bounds, high=high_bounds)

# Create the SBI inference object
inference = SNPE(prior=prior)

# Generate the parameter sets (theta) from the stable prior
theta = prior.sample((10000,))

# Run the simulations. With the stable prior, there should be no NaNs.
x = []
for th in theta:
    x.append(ar1_simulator(th))
x = torch.stack(x)

# Append the simulations to the inference object
inference.append_simulations(theta=theta, x=x)

# Train the neural network
density_estimator = inference.train()

# Build the posterior distribution
posterior = inference.build_posterior(density_estimator)

# ---

# 5. ANALYZE AND PLOT RESULTS
posterior_samples = posterior.sample((5000,), x=observations)

from sbi import analysis as analysis
fig, axes = analysis.pairplot(
    posterior_samples, 
    limits=[[-1, 1], [0, 2]], 
    labels=[r'$\phi$', r'$\sigma$']
)
plt.show()
```
